```python
import folium
import folium.plugins as plugins
import numpy as np
import pandas as pd
```

```python
#Ritis Incident Data
```

```python
infile2=r'C:\On going Research Works\Safety Project\Data\Ritis_Data\ritis-events_I75.csv'

df2=pd.read_csv(infile2,parse_dates=['Time Opened','Time Closed'],usecols=['Standardized Type','County','Latitude','Longitude','Time Opened','Time Closed','Duration'])

df2=df2[(df2['Standardized Type']=='Collision')]

df2['dates']=df2['Time Opened'].dt.date

df2=df2.sort_values(by=['dates'])

df2['popup']="Standardized Type: " + df2['Standardized Type'] +' & '+ "Duration: "+ df2['Duration']

date2=list((df2['dates']).unique())

gn2=date2

df2=df2.dropna(how='any')

df2=df2.reset_index()

data2=[]

for i in range (len(date2)):
    print(date2[i])
    a=df2[(df2['dates']==date2[i])]
    b=[[round(float(row['Latitude']),4),round(float(row['Longitude']),4),1] for index, row in a.iterrows()]
    data2.append(b)
    
time_index2=[date2[n].strftime('%m-%d-%Y') for n in range (len(date2))]
```

```python
import folium

m = folium.Map([27.0, -81.0], zoom_start=6)

hm = plugins.HeatMapWithTime(
    data2,
    index=time_index2,
    name='Dynamic Heatmap',
    radius=7,
    gradient={0.2: 'blue', 0.4: 'lime', 0.6: 'orange', 1: 'red'},
    min_opacity=0.5,
    max_opacity=1.0,
    scale_radius=False,
    use_local_extrema=True,
    auto_play=True,
    display_index=True,
    index_steps=1,
    min_speed=0.1,
    max_speed=10,
    speed_step=0.1,
    position='bottomleft',
    overlay=True,
    control=True,
    show=True,
)

hm.add_to(m)

plugins.Fullscreen(
    position='topright',
    title='Expand me',
    title_cancel='Exit me',
    force_separate_button=True).add_to(m)

legend_html =   '''
                <div.a style="position: fixed; 
                            bottom: 25px; right: 20px; width: 350px; height: 90px; 
                            border:2px; z-index:9999; font-size:18px;font-weight:bold; color: blue; text-align: center;">
                            Safety Analytics and Visulization Engine 
                                  
                </div.a>
                
                <div.b style="position: fixed; 
                            bottom: 5px; right: 40px; width: 300px; height: 90px; 
                            border:2px; z-index:9999; font-size:14px; color: blue; text-align: center; text-decoration: underline">
                            Developed by UNMD LAB, UCF
                </div.a>
                </div.b>
                ''' 
m.get_root().html.add_child(folium.Element(legend_html))

m
```

```python
fg1 = folium.FeatureGroup(name='Dynamic Heatmap')

fg2 = folium.FeatureGroup(name='Hurricane Irma Incident data Ritis')
```

```python
mcg2 = folium.plugins.MarkerCluster(name='Hurricane Irma Incident data Ritis ',control=True).add_to(fg2)   # Marker Cluster, hidden in controls

for k in range(len(date2)):
    df_sample2=df2[df2['dates']==gn2[k]]
    data2=[[row2['Latitude'],row2['Longitude']] for index2, row2 in df_sample2.iterrows()]
    gn2[k] = folium.plugins.FeatureGroupSubGroup(mcg2, str(date2[k]))
    for l in range (len(data2)):
        gn2[k].add_child(folium.Marker(data2[l],popup=str(df2['popup'][l])))
    m.add_child(mcg2)
    m.add_child(gn2[k])
    

folium.LayerControl(collapsed=True).add_to(m)
```
